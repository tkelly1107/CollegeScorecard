Load in Recent Cohort Field of Study data

```python
df = pd.read_csv('/Most-Recent-Cohorts-Field-of-Study1.csv')
```

```python
df
```

Select for Masters, Bachelors, Professional and Doctoral Degree programs.

```python
# Filter the DataFrame to only include rows where CREDLEV is 3, 5, 6, or 7
filtered_df = df[df['CREDLEV'].isin([3, 5, 6, 7])]

# Display the filtered DataFrame
filtered_df
```

Remove For Profit and Foreign schools.

```python
# Filter out "Private, for-profit" in the CONTROL column
filtered_df = filtered_df[filtered_df['CONTROL'] != 'Private, for-profit']
# Filter out "Foreign" in the CONTROL column
filtered_df = filtered_df[filtered_df['CONTROL'] != 'Foreign']
filtered_df
```

Filter for main campuses only.

```python
filtered_df = filtered_df[filtered_df['MAIN'] == 1]
filtered_df
```

Filter for field of study Medical Informatics.

```python
troubleshooting = filtered_df[filtered_df['CIPDESC'].str.contains('Medical Illustration and Informatics', case=False)]
troubleshooting
```

```python
filtered_df = filtered_df[filtered_df['CIPDESC'] == 'Medical Illustration and Informatics.']
filtered_df
```

Filter for 1 in DISTANCE column to select for residential programs.

```python
residential = filtered_df[filtered_df['DISTANCE'] == 1]
residential
```

Select for online programs only.

```python
online = filtered_df[filtered_df['DISTANCE'] == 3]
online
```

Select for hybrid programs only.

```python
hybrid = filtered_df[filtered_df['DISTANCE'] == 2]
hybrid
```

Save each dataframe.

```python
import os

# Create a new folder in content
new_folder_path = '/content/collegescorecard_projectdata'
os.makedirs(new_folder_path)

# Save the CSV files to the new folder
online.to_csv(os.path.join(new_folder_path, 'online.csv'))
hybrid.to_csv(os.path.join(new_folder_path, 'hybrid.csv'))
residential.to_csv(os.path.join(new_folder_path, 'residential.csv'))
```
